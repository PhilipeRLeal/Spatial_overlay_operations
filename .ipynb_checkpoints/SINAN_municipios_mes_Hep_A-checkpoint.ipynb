# Operações sobre dados do SINAN Município de residência por Mês entre 2007 e 2017:

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
import geopandas as gpd
```

```python
Path = r'C:\Doutorado\Tese\SINAN\Casos_hepatite_A_por_municipio_por_mes_ano'

File_names = {}

for root, dirs, files in os.walk(Path):
    for name in files:
        if name.endswith((".csv", ".xlsx")):
            File_names[str(name.split('.')[0])] = os.path.join(str(root), str(name))

File_names.pop('DATASET_FULL_YEARS', None)            
print(File_names['2007'])
```

```python
File_names.keys()
```

```python
def Month_converter(String):
    Months = {'jan': '01',
              'fev': '02',
             'mar':'03',
             'abr': '04',
             'mai': '05',
             'jun': '06',
             'jul':'07',
             'ago':'08',
             'set':'09',
             'out':'10',
             'nov':'11',
             'dez':'12'}
    String = np.str.lower(str(String))
    
    return Months[String[0:3]]

Month_converter('Mai')
```

```python
Datasets = []

for Year in File_names.keys():
    print('\n\n', Year)
    try:
        Table = pd.read_csv(File_names[str(Year)], header=[4] , encoding = 'latin', skipfooter=29, engine='python', na_values='-')
    except:
        Table = pd.read_csv(File_names[str(Year)], header=[4] , encoding = 'latin', engine='c', na_values='-')

    
    
    Table.drop('Total', axis=1, inplace=True)


    Meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']

    Ds = pd.melt(Table,value_vars=Meses, id_vars = ['Município de residência'], var_name='Mes', value_name='N_Hepat_A_por_Mes')
    
    Ds.rename({'Município de residência':'Municipio_de_residencia'}, axis=1, inplace=True)
    
    Ds['Mes'] = Ds['Mes'].apply(np.str.lower)
    
    Ds['Mes'] = Ds['Mes'].apply(Month_converter)
    Ds['DATETIME'] = str(Year) + '/' + Ds['Mes']

    Ds.drop('Mes', axis=1, inplace=True)
    
    print(Ds.head())
    
    Datasets.append(Ds)

    
```

```python
Dataset = pd.concat(Datasets, axis=0)

Dataset.head(5)
```

```python
def Get_IBGE_CODE (STRING, N=0):
    if N==0:
        IBGE_CODE = STRING.split(' ')[N]
        return IBGE_CODE
    if N!= 0:
        IBGE_CODE = STRING.split(' ')[N:]
    return ' '.join(IBGE_CODE)

print(Get_IBGE_CODE('Abc 234 ABC 000', 0))
print(Get_IBGE_CODE('Abc 234 ABC 000', 1))
```

# Aula 4: regressão Estatística - Auto correlação

    # Auto correlação negativa: mean reverted - com o passar do tempo, o valor diminui.
    
    # Auto correlação positiva: trend following - com o passar do tempo, o valor aumenta

```python
# autocorrelação temporal:
Auto_corr = []

for i in range(1,7):
    print(i)
    Multiplier = str(i * 6) + 'M'

    
    Parah_Resampled = Parah.resample(rule=Multiplier).last()
    Autocorr_Parah_percentage = Parah_Resampled['N_Hepat_A_por_Mes'].pct_change().autocorr()
    Autocorr_Parah = Parah_Resampled['N_Hepat_A_por_Mes'].autocorr()
    Multiplier = i * 6
    Auto_corr.append([Multiplier, float(Autocorr_Parah_percentage), float(Autocorr_Parah)])
    
Auto_corr = np.array(Auto_corr)

```

```python
Auto_corr = pd.DataFrame(Auto_corr)

Auto_corr.columns=['Lag Mensal',  'Autocorr_PCT_CHANGE', 'Autocorr']

Auto_corr.head()
```

```python
Auto_corr.set_index('Lag Mensal', inplace=True)

```

```python
fig, Ax = plt.subplots()

Auto_corr['Autocorr_PCT_CHANGE'].plot(kind='line', c='r', legend=True)
Ax.set_title("Figura de autocorrelação \npor Percentage Change")

Auto_corr['Autocorr'].plot(kind='line', c='b', legend=True)
Ax.set_title("Figura de autocorrelação")

plt.show()

fig.savefig(os.path.join(Path, 'Figura de autocorrelação.png'))
```
