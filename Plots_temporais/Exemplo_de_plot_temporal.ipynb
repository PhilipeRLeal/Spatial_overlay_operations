```python
def Get_Prevalencia(Instancia):
    
    HAV = Instancia['HAV'] *1.0
    
    POP_ESTIMADO = Instancia['POP_ESTIMADO']
    
    Prevalencia = HAV/POP_ESTIMADO
    
    if np.isnan(Prevalencia):
         Prevalencia = 0.0
    return Prevalencia
    
```

```python
Merged_SHP_POP_HAV['Prevalencia'] = Merged_SHP_POP_HAV.apply(Get_Prevalencia, axis=1)
```

# Fazendo Multiplos Plots temporais:

```python
import matplotlib as mpl
cmap = mpl.cm.hot_r

def GDF_multiplots_base_function(fig, Axes, GDF, Attribute_name, N=5):
    
    
    Vmin = GDF[str(Attribute_name)].min()
    Vmax = GDF[str(Attribute_name)].max()
    if Vmax == Vmin:
        Vmax = 100
    
    Counter = 10
    while np.floor(np.round(Vmax, 1)) == 0:
        Counter = Counter *10 
        Vmax = Vmax * Counter
    
    
    Ticks_list, step = np.linspace(Vmin, Vmax, num=N, endpoint=True, retstep=True)
    Ticks_list = np.round(Ticks_list,2)
    
    sm = plt.cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(vmin=Vmin-(step/2),vmax=Vmax+(step/2)))

    sm._A = []
    
    cbar = fig.colorbar(sm, ax=Axes, ticks=Ticks_list)   
    cbar.ax.tick_params(labelsize=6) 
    cbar.set_label('X ' + str(Counter), fontsize=9)
    #ticks = [t.get_text().replace('.',',')+'%' for t in cbar.ax.get_yticklabels()]
    
    #cbar.set_ticklabels(ticks)
    
    return Counter
```

```python
fig, ax = plt.subplots(3,4, sharex=True, sharey=True)
ax = ax.ravel()
for k, i in enumerate(range(2008,2018)):
    axes = ax[k]
    
    
    # Plotando
    
    
    Merged_SHP_POP_HAV.loc[Merged_SHP_POP_HAV.Datetime == dt.datetime(i,12,31)].plot(ax=axes,cmap=cmap, column='Prevalencia', label='Prevalencia')
    Merged_SHP_POP_HAV.loc[Merged_SHP_POP_HAV.Datetime == dt.datetime(i,12,31)].plot(ax=axes, facecolor='white', alpha=0.2, edgecolor='k', linewidth=0.3)
    
    
    
    
    # Calculando limites para o Colorbar:

    GDF_multiplots_base_function(fig, axes, Merged_SHP_POP_HAV.loc[Merged_SHP_POP_HAV.Datetime == dt.datetime(i,12,31)], 'Prevalencia')
    
    # Titulo mais counter
    
    axes.set_title(str(dt.datetime(i,12,31).year), fontsize=10)
    
    AXES_YTICKS = axes.get_yticks()
    
    axes.set_yticklabels(AXES_YTICKS, fontsize=6)
    axes.set_xticklabels(AXES_YTICKS, fontsize=6)
    
fig.subplots_adjust(top=0.95,
bottom=0.055,
left=0.065,
right=0.94,
hspace=0.26,
wspace=0.403)


import matplotlib.patches as mpatches

x_tail = 0.975
y_tail = 0.01
x_head = 0.975
y_head = 0.035
dx = x_head - x_tail
dy = y_head - y_tail

arrow = mpatches.FancyArrowPatch((x_tail, y_tail), (x_head, y_head), dpi_cor=1200  , transform=fig.transFigure, color='k')


arrow = mpatches.Arrow(x_tail, y_tail, dx, dy, width=0.05, transform=fig.transFigure, color='k', figure=fig)

fig.patches.extend([arrow])

plt.text(x_head, y_head + 0.8*(y_head - y_tail) , s='N', size=11, ha='center', va='center',
                 color='K', transform=fig.transFigure, figure=fig)

plt.show()
```

```python
def Geometry_to_STR(GDF):
    DF = pd.DataFrame(GDF).copy()
    
    DF.geometry = DF.geometry.apply(lambda x: str(x))
    
    return DF

def STR_Geometry_to_Geometry(DF):
    GDF = gpd.GeoDataFrame(DF).copy()
    from shapely.wkt import loads
    GDF.geometry = GDF.geometry.apply(loads)
    
    return GDF

def Groupby_GDF(GDF, Lista_de_grupos, Lista_de_atributos_numericos, lista_de_funcoes_de_reducao):  
    
    DF = Geometry_to_STR(GDF)
    print("Transformação da Geometria em formato str concluida!")
    
    DF_GROUPED = DF.groupby(Lista_de_grupos)[Lista_de_atributos_numericos].agg(lista_de_funcoes_de_reducao)
    
    DF_GROUPED.reset_index(inplace=True)
    
    print("\n\nAgrupamento realizado com sucesso!\n\n")
    
    
    GDF = STR_Geometry_to_Geometry(DF_GROUPED)
    
    print("Transformação da geometria em formato Shapely bem sucedida!")
    
    return GDF

GDF = Groupby_GDF(Merged_SHP_POP_HAV, ['COD_MUNIC_6', 'geometry'], ['Prevalencia'], [np.mean, np.sum])

print(GDF.head())


GDF.columns.get_level_values(0).unique()

COLLUMNS = GDF.columns.get_level_values(0).unique()
Group_functions = GDF.columns.get_level_values(1).unique()[1:]
COLLUMNS = COLLUMNS[:-1]

COLLUMNS = COLLUMNS.append(Group_functions)



GDF.columns = GDF.columns.droplevel()
GDF.columns = COLLUMNS
GDF.head()




```
